```python
from dynamical_systems.mass_spring_damper_system import MassSpringDamperEnv
from random_exploration import random_exploration

import numpy as np
import matplotlib.pyplot as plt

env = MassSpringDamperEnv(nlin=True)

# HYPERPARAMETERS
NUM_TRAJECTORIES = 1
HORIZON = 500

states, _, _ = random_exploration(env, NUM_TRAJECTORIES, HORIZON)
states = np.array(states)  # Convert list of states to a NumPy array
time_steps = np.arange(len(states))  # Time steps (x-axis)

# Plot each dimension of the state
plt.figure(figsize=(10, 6))
plt.plot(time_steps, states[:, 0], label="Position x", color="blue")
plt.plot(time_steps, states[:, 1], label="Velocity v", color="red")

plt.xlabel("Time Steps")
plt.ylabel("State Values")
plt.title("Spring-Mass-System")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import gymnasium as gym
from random_exploration import random_exploration

import numpy as np
import matplotlib.pyplot as plt

env = gym.make("CartPole-v1")

# HYPERPARAMETERS
NUM_TRAJECTORIES = 10
HORIZON = 100

states, _, _ = random_exploration(env, NUM_TRAJECTORIES, HORIZON)
states = np.array(states)  # Convert list of states to a NumPy array
time_steps = np.arange(len(states))  # Time steps (x-axis)

# Plot each dimension of the state
plt.figure(figsize=(10, 6))
plt.plot(time_steps, states[:, 0], label="Position x", color="blue")
plt.plot(time_steps, states[:, 1], label="Velocity v", color="green")
plt.plot(time_steps, states[:, 2], label="Angle θ", color="orange")
plt.plot(time_steps, states[:, 3], label="Angle Vel ω", color="red")

plt.xlabel("Time Steps")
plt.ylabel("State Values")
plt.title("Cartpole")
plt.legend()
plt.grid(True)
plt.show()
```
